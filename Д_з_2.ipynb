```python
x = [1, 3, 7]
y = [2, 6, 14]
Y = []
w0 = 0
w1 = 0
lr = 0.01
for i in range(len(x)):
  prediction = w0 + w1* x[i]
  w1 += 2 * lr * x[i]* (y[i] - prediction)
  w0 += 2 * lr * (y[i] - prediction)
print(f'w0 = {w0}, w1 = {w1}')

for j in x:
  Y.append(w0 + w1*j)
```

```python
import matplotlib.pyplot as plt
plt.plot(x, Y, color='red')
plt.scatter(x, y, color='green')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

**Вывод**: получились хорошие веса и они похожи на "настоящие"
