```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def function (x):
  return (np.sin(x) + np.cos(x))**2
X = np.linspace(0, 11, 100)
Y = function(X)
plt.plot(X, Y, color='g',label='Вот такая моя функция', dashes=[8,4], alpha=0.5)
plt.grid()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('y = (sin(x) + cos(x))**2')
plt.legend()
plt.show()
```

```python
X = np.random.normal(0, 1, 3000)
Y = np.random.normal(3, 4, 3000)
plt.scatter(X, Y, s=10, color='b',alpha=0.6, marker='<')
plt.xlabel('Ось Х')
plt.ylabel('Ось Y')
plt.grid()
plt.title('Точки')
plt.show()
```
